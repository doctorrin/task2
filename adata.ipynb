```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup

a = []
b = []
c = []
d = []
e = []

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'}

response = requests.get('https://www.goszakup.gov.kz/ru/registry/rqc', headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
soup = soup.find('div', class_='table-responsive')

for i in soup.find_all('a'):
    a.append(i.text.strip())
    response2 = requests.get(i['href'])
    soup2 = BeautifulSoup(response2.text, 'html.parser')
        
    head = soup2.find_all('table', class_ = 'table table-striped')[2]
    d.append((head.find_all('td')[0]).text.strip())
    c.append((head.find_all('td')[2]).text.strip())

    
    address = soup2.find_all('table', class_ = 'table table-striped')[3]
    e.append((address.find_all('td')[3]).text.strip())
    
for i in range(2, len(soup.find_all('td')), 4):
    b.append(soup.find_all('td')[i].text.strip())

        
D = {}    

D['Наименование организации'] = a
D['БИН организации'] = b
D['ФИО руководителя'] = c
D['ИИН руководителя'] = d
D['Полный адрес организации'] = e

df = pd.DataFrame.from_dict(D)
#
df = df.drop_duplicates()
df
```
